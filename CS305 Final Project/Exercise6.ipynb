<P>To start, we'll read in the data and store them in <code>numpy</code> arrays <code>X</code> and <code>y</code>. If we were reading in 32x32 grayscale images, we might have 32x32 features, one for each pixel. Here, we have color images and there are three values for each pixel corresponding to the RGB (red, green, blue) color associated with the pixel.</P>

```python
# Read in data and store in numpy arrays X and y
import numpy as np
import pickle
with open('cifar', 'rb') as fo: DATA = pickle.load(fo, encoding='bytes')
X = np.array(DATA[b'data'], dtype=float) / 255.0
y = np.array(DATA[b'labels'])
```

<font color="maroon"><u>How many images are there in the dataset? How many features does each image have?</u></font>

<P>Just for fun, let's plot 50 of the images. Each image below is labeled with a digit, 0 through 9, indicating which of the ten classes it corresponds to.</P>

```python
# Plot many small images
import matplotlib.pyplot as plt
X = X.reshape([-1, 3, 32, 32])
X = X.transpose([0, 2, 3, 1])
numImages, perRow = 50, 10
f, axarr = plt.subplots(int(numImages/perRow), perRow, figsize=(8,8))
for i in range(numImages):
    image = X[-numImages+i,:,:,:]
    axarr[int(i/perRow),i%perRow].imshow(image)
    axarr[int(i/perRow),i%perRow].set_title(str(int(y[-numImages+i])))
    plt.setp(axarr[int(i/perRow),i%perRow].get_xticklabels(), visible=False)
    plt.setp(axarr[int(i/perRow),i%perRow].get_yticklabels(), visible=False)
    axarr[int(i/perRow),i%perRow].axis('off')
plt.tight_layout()
plt.show()
X = X.reshape((X.shape[0], X.shape[1]*X.shape[2]*X.shape[3]))
```
