# Data Visualization 

```python
import numpy as np 
import pandas as pd
```

```python
DATA = pd.read_csv('clean_data/tracks.csv', skiprows=0, header=0)
DATA = pd.DataFrame(DATA).drop(columns=['id'])
```

```python
features = list(DATA.columns)
```

```python
hits = np.array(DATA.loc[DATA['hit']==1])
non_hits = np.array(DATA.loc[DATA['hit']==0])
```

## Featurization 

In this step, I will extract each feature.

```python
def get_features(DATA, features_list):
    features_vals = {}
    for i in range(len(features_list)):
        feature = features_list[i]
        features_vals[feature] = DATA[:,i]
    return features_vals
```

```python
hits_features = get_features(hits, features)
not_hits_features = get_features(non_hits, features)
```

## Hit Songs Distribution

```python
import matplotlib.pyplot as plt
from matplotlib import colors


fig = plt.figure()
plt.hist(hits_features['tempo'], bins=15)
plt.title('Tempo Distribution Hit Songs')
plt.xlabel('tempo')
plt.ylabel('# of Songs')
fig.savefig('tempo_hit.png', dpi=fig.dpi)
plt.show()
```

```python
plt.hist(hits_features['danceability'], bins=15)
plt.title('Danceability Distribution Hit Songs')
plt.xlabel('danceability')
plt.ylabel('# of Songs')
fig.savefig('danceability_hit.png', dpi=fig.dpi)
plt.show()
```

```python
plt.hist(hits_features['energy'], bins=15)
plt.title('Energy Distribution Hit Songs')
plt.xlabel('energy')
plt.ylabel('# of Songs')
fig.savefig('energy_hit.png', dpi=fig.dpi)
plt.show()
```

```python
plt.hist(hits_features['mode'], bins=2)
plt.title('Mode Distribution Hit Songs')
plt.xlabel('mode')
plt.ylabel('# of Songs')
fig.savefig('mode_hit.png', dpi=fig.dpi)
plt.show()
```

```python
plt.hist(hits_features['track_popularity'], bins=15)
plt.title('Song Popularity Distribution Hit Songs')
plt.xlabel('popularity')
plt.ylabel('# of Songs')
fig.savefig('track_hit.png', dpi=fig.dpi)
plt.show()
```

```python
plt.hist(hits_features['artist_popularity'], bins=15)
plt.title('Artist Popularity Distribution Hit Songs')
plt.xlabel('popularity')
plt.ylabel('# of Songs')
fig.savefig('artist_hit.png', dpi=fig.dpi)
plt.show()
```

```python
plt.scatter(hits_features['artist_popularity'], hits_features['track_popularity'], c='green', alpha=0.7)
plt.title('Track and Artist Popularity Plot')
plt.xlabel('Artists')
plt.ylabel('Tracks')
fig.savefig('artist_track_hit.png', dpi=fig.dpi)

plt.show()
```

```python
plt.hist(hits_features['liveness'],bins=15)
plt.title('Liveness Distribution Hit Songs')
plt.xlabel('liveness')
plt.ylabel('# of Songs')
fig.savefig('liveness_hit.png', dpi=fig.dpi)
plt.show()
```

```python
plt.hist(hits_features['speechiness'], bins=15)
plt.title('Speechiness Distribution Hit Songs')
plt.xlabel('speechiness')
plt.ylabel('# of Songs')
fig.savefig('speechiness_hit.png', dpi=fig.dpi)
plt.show()
```
