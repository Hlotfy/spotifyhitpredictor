# Data Visualization 

```python
import numpy as np 
import pandas as pd
```

```python
DATA = pd.read_csv('clean_data/even_more_tracks.csv', skiprows=0, header=0)
DATA = pd.DataFrame(DATA).drop(columns=['id'])
```

```python
features = list(DATA.columns)
```

```python
hits = np.array(DATA.loc[DATA['hit']==1])
non_hits = np.array(DATA.loc[DATA['hit']==0])
```

## Featurization 

In this step, I will extract each feature.

```python
def get_features(DATA, features_list):
    features_vals = {}
    for i in range(len(features_list)):
        feature = features_list[i]
        features_vals[feature] = DATA[:,i]
    return features_vals
```

```python
hits_features = get_features(hits, features)
not_hits_features = get_features(non_hits, features)
```

```python
DATA = pd.read_csv('clean_data/track_data.csv', skiprows=0, header=0)
DATA = pd.DataFrame(DATA).drop(columns=['id', 'year'])
features = list(DATA.columns)
```

```python
DATA = np.array(DATA)

X = DATA[:, :-1]
y = DATA[:,-1]
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)
X= scaler.transform(X)
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(20)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(random_state=0)
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
clf.score(X_train, y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix

y_pred = clf.predict(X_test)
tn, fp, fn, tp = confusion_matrix(y_test,y_pred).ravel()
print("true negative: ", tn, "\nfalse positive: ", fp, "\nfalse negative: ", fn, "\ntrue positive: ", tp)
print(classification_report(y_test, y_pred))
```

```python
y_pred = clf.predict(X_train)
tn, fp, fn, tp = confusion_matrix(y_train, y_pred).ravel()
print("true negative: ", tn, "\nfalse positive: ", fp, "\nfalse negative: ", fn, "\ntrue positive: ", tp)
print(classification_report(y_train, y_pred))
```

```python
features = features[:-1]
print(features)
featDict = {}
for i in range(len(features)):
    feature = features[i]
    featDict[feature] = clf.coef_[0][i]
```

```python
import operator

sorted(featDict.items(), key=operator.itemgetter(1))
```

```python
from sklearn import tree
from sklearn.tree.export import export_graphviz

clf = tree.DecisionTreeClassifier()

clf.fit(X_train, y_train)

out = export_graphviz(clf, out_file='tree.dot')
feat_importance = clf.tree_.compute_feature_importances(normalize=False)
print("feat importance = " + str(feat_importance))

features = features[:-1]
featDict = {}
for i in range(len(features)):
    feature = features[i]
    featDict[feature] = feat_importance[i]

sorted(featDict.items(), key=operator.itemgetter(1))
```

```python
clf.score(X_test, y_test)
```

```python
y_pred = clf.predict(X_test)
tn, fp, fn, tp = confusion_matrix(y_test,y_pred).ravel()
print("true negative: ", tn, "\nfalse positive: ", fp, "\nfalse negative: ", fn, "\ntrue positive: ", tp)
print(classification_report(y_test, y_pred))
```
